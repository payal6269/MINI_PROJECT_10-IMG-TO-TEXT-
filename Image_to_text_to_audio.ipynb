```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/Picture1.png')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
tts = gtts.gTTS(result)
tts.save("hello.mp3")
```

```python
# 🛠️ Step 1: Install required libraries
!pip install pytesseract gTTS Pillow --quiet
!apt install tesseract-ocr -y > /dev/null

# 📚 Step 2: Import all required libraries
from PIL import Image
import pytesseract
from gtts import gTTS
from IPython.display import Audio
from google.colab import files


# 🔍 Step 4: Extract text
text = pytesseract.image_to_string(img)
print("📝 Extracted Text:\n", text)

# 🗣️ Step 5: Convert to audio if text exists
if text.strip():
    tts = gTTS(text)
    tts.save("output.mp3")
    print("✅ Audio saved as output.mp3")

    # ▶️ Step 6: Play the audio
    display(Audio("output.mp3"))

    # 📥 Step 7: Download the audio file
    files.download("output.mp3")
else:
    print("❌ No text found in the image.")
```
